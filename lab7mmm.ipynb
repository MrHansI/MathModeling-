# 3. Математическая постановка задачи.
Задача состоит в исследовании движения вентилятора массой $M_fan$, подвешенного на пружине длиной $L$, на фоне ускорения свободного падения $g$. Вентилятор вращается с периодом $T_{fan}$, а пружинный маятник колеблется с периодом $T$.

Уравнение движения вентилятора и пружинного маятника можно записать в виде дифференциальных уравнений, учитывая законы сохранения массы, импульса и энергии:

1. Закон сохранения импульса вдоль оси вращения вентилятора: $M_{\text{fan}} \cdot \omega_{\text{fan}} \cdot L = m \cdot \omega \cdot L$ 


2. Закон сохранения энергии: $M_{\text{fan}} \cdot g \cdot h = \frac{1}{2} \cdot I \cdot \omega_{\text{fan}}^2 + \frac{1}{2} \cdot m \cdot v^2$

3. Уравнение движения пружинного маятника: $\frac{d^2\theta}{dt^2} = -\frac{g}{L} \cdot \theta$
Где $\omega_fan$ - угловая скорость вентилятора, $\omega$ - угловая скорость маятника, $I$ - момент инерции вентилятора, $h$ - высота подъема центра масс вентилятора.

4. Для получения фигуры Лиссажу, необходимо найти соотношение между периодами $T_{fan}$ и $T$, при котором траектория концов лопастей вентилятора будет иметь форму фигуры Лиссажу. Это можно сделать путем нахождения таких значений $T_{fan}$ и $T$, при которых угол отклонения маятника $\theta$ зависит от угла вращения вентилятора $\phi$ так, чтобы форма кривой была фигурой Лиссажу.

5. Математически задачу можно решить численными методами интегрирования дифференциальных уравнений системы, учитывая все указанные параметры системы, и анализировать траекторию движения концов лопастей вентилятора для различных соотношений периодов $T_{fan}$ и $T$.

Таким образом, математическая постановка задачи заключается в нахождении соотношения периодов вращения вентилятора и колебаний маятника, при котором траектория концов лопастей вентилятора будет представлять фигуру Лиссажу, и решении системы дифференциальных уравнений, учитывая все параметры системы.






```python
omega_fan = 2 * np.pi  # Угловая скорость вращения лопастей вентилятора
omega_spring = 2 * np.pi  # Угловая скорость пружинного маятника 


t = np.linspace(0, 10, 1000)


theta = np.sin(omega_fan * t)
phi = np.sin(omega_spring * t)


plt.figure(figsize=(8, 6))
plt.plot(theta, phi)
plt.xlabel('Угол вентилятора (θ)')
plt.ylabel('Угол маятника (φ)')
plt.title('Траектория концов лопастей вентилятора в виде фигуры Лиссажу')
plt.grid(True)
plt.show()
```
