# Математическая постановка задачи 

## <center > Пункт 1: <center> 




### <center> Зависимость времени затухания колебаний маятника от жесткости пружины (k)< center> 



Дано:




* Математический маятник с массой M.
* Пружина с жесткостью k.
* Уравнение движения математического маятника в вязкой среде:
<center> $ M\frac{d^2x}{dx^2} + c\frac{dx}{dt} + kx = 0$ <center> 
    
    
* Желаемая зависимость: времени затухания колебаний T от жесткости пружины k.    
    
    
    
1.   Перепишем уравнение движения в стандартной форме для гармонических колебаний:
    
    
<center> $ M\frac{d^2x}{dx^2} + c\frac{dx}{dt} + kx = 0$ <center>
    
    
2. Предположим, что решение имеет вид:
    
    
    
<center> $x(t) = A\exp ^{rt} $ <center> 
    
    
где A - амплитуда колебаний, r - характеристический корень.
    
    

3. Подставим это предположение в уравнение движения и получим характеристическое уравнение:
    
    

<center> $ Mr^2 + cr + k =0 $ <center> 
    
    

4. Решим характеристическое уравнение для r и найдем два значения r1 и r2:
    
    
<center> $ r_{1,2} = \frac{-c +- \sqrt{c^2 - 4Mk}}{2M} $ <center> 
    
    
    
5. Время затухания колебаний T связано с коэффициентами затухания вида c/2M и угловой частотой ω0 вида sqrt(k/M) следующим образом:
    
    
    
<center> $ T = \frac{2}{r_1} =  \frac{2}{r_2} =  \frac{2}{c/2M} =  \frac{4M}{c} $ <center> 
    
Таким образом, временем затухания T зависит обратно пропорционально от коэффициента сопротивления c. 
    
    
    
    
## Пункт 2 
    
    
### Зависимость времени затухания колебаний маятника от его массы (M)

Дано:

* Математический маятник с жесткостью пружины k.
* Пружина с жесткостью k.
* Уравнение движения математического маятника в вязкой среде:
    
    
    
    
<center> $ M\frac{d^2x}{dx^2} + c\frac{dx}{dt} + kx = 0$ <center>
    
    
    
    
Желаемая зависимость: времени затухания колебаний T от массы маятника M.
    
    
    
1. Уравнение движения математического маятника:
    
    
<center> $ M\frac{d^2x}{dx^2} + c\frac{dx}{dt} + kx = 0$ <center>
    
    
    
2. Предполагаем, что решение имеет вид:
    
    
    
<center> $x(t) = A\exp ^{rt} $ <center> 
    
    
где A - амплитуда колебаний, r - характеристический корень.
    
    

3. Подставим это предположение в уравнение движения и получим характеристическое уравнение:
    
    

<center> $ Mr^2 + cr + k =0 $ <center> 
    
    

4. Решим характеристическое уравнение для r и найдем два значения r1 и r2:
    
    
<center> $ r_{1,2} = \frac{-c +- \sqrt{c^2 - 4Mk}}{2M} $ <center> 
    
    
    
    
5. Время затухания колебаний T связано с коэффициентами затухания вида c/2M и угловой частотой $ω_0$ вида  $ \sqrt{\frac{k}{M}} $  следующим образом:


    
<center> $ T = \frac{2\pi}{w_0} = \frac{2\pi}{\sqrt{\frac{k}{M}}} = \frac{2\pi}{\sqrt{M*k}} $ <center> 
    
    
    
6. Таким образом, время затухания T зависит обратно пропорционально квадратному корню из произведения массы маятника M и жесткости пружины k:
    
    
<center> $ T = \frac{2\pi}{\sqrt{M*k}} $ <center>
    
    
    
Эта зависимость демонстрирует, что время затухания увеличивается с уменьшением массы маятника M или увеличением жесткости пружины k.

## Пункт 4
    
    
###  Зависимость периода колебания маятника от коэффициента сопротивления и определение, при каком коэффициенте сопротивления колебания маятника отсутствуют

Дано:

* Математический маятник с жесткостью пружины k.
* Пружина с жесткостью k.
* Уравнение движения математического маятника в вязкой среде:
    
    
    
    
<center> $ M\frac{d^2x}{dx^2} + c\frac{dx}{dt} + kx = 0$ <center>
    
    
    
    
Желаемая зависимость: периода колебания маятника T от коэффициента сопротивления c, и определение значения коэффициента сопротивления c, при котором колебания маятника отсутствуют.
    
    
1. Уравнение движения математического маятника:
    
    
<center> $ M\frac{d^2x}{dx^2} + c\frac{dx}{dt} + kx = 0$ <center>
    
    
    
2. Предполагаем, что решение имеет вид:
    
    
    
<center> $x(t) = A\exp ^{rt} $ <center> 
    
    
где A - амплитуда колебаний, r - характеристический корень.
    
    

3. Подставим это предположение в уравнение движения и получим характеристическое уравнение:
    
    

<center> $ Mr^2 + cr + k =0 $ <center> 
    
    

4. Решим характеристическое уравнение для r и найдем два значения r1 и r2:
    
    
<center> $ r_{1,2} = \frac{-c +- \sqrt{c^2 - 4Mk}}{2M} $ <center>
    
    
5. Период колебания маятника T связан с угловой частотой ω0 и коэффициентом сопротивления c следующим образом:
    
    
<center> $ T = \frac{2\pi}{w_0} = \frac{2\pi}{\sqrt{\frac{k}{M} - (\frac{c}{2M})^2}} $ <center> 
    
    
6. Для отсутствия колебаний (т.е., амплитуда колебаний становится равной нулю), необходимо, чтобы коэффициент сопротивления c превысил критическое значение $c_{crit} $, при котором знаменатель под корнем в уравнении для T становится отрицательным. Это означает, что колебания маятника отсутствуют, когда:
    
    
<center> $ c > c_{crit} = 2\sqrt{Mk} $ <center> 
    
    
    
Таким образом, зависимость периода колебания T от коэффициента сопротивления c представляется формулой, и определено значение $ c_{crit} $ , при котором колебания маятника отсутствуют

# Реализация : 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

# Пункт 1: Зависимость времени затухания колебаний маятника от жесткости пружины (k)

```python
M = 1.0  # Масса маятника
c = 1.0  # Коэффициент вязкого трения
k_values = np.linspace(1, 20, 100)  

T_values = 2 * np.pi / np.sqrt(k_values / M - (c / (2 * M))**2)


plt.figure(figsize=(8, 6))
plt.plot(k_values, T_values, label='T(k)')
plt.xlabel('Жесткость пружины k')
plt.ylabel('Время затухания T (сек)')
plt.title('Зависимость времени затухания от жесткости пружины маятника')
plt.legend()
plt.grid(True)
plt.show()






```

# Пункт 2: Зависимость времени затухания колебаний маятника от его массы (M)

```python
k = 10.0  # Жесткость пружины
c = 1.0   # Коэффициент вязкого трения
M_values = np.linspace(0.1, 1.0, 100)  # Диапазон масс маятника

T_values = 2 * np.pi / np.sqrt(k * M_values)

plt.figure(figsize=(8, 6))
plt.plot(M_values, T_values, label='T(M)')
plt.xlabel('Масса маятника (кг)')
plt.ylabel('Время затухания T (сек)')
plt.title('Зависимость времени затухания от массы маятника')
plt.legend()
plt.grid(True)
plt.show()
```

# Пункт 3: Зависимость времени затухания колебаний маятника от начального отклонения (α)

```python
k = 10.0  # Жесткость пружины
c = 1.0   # Коэффициент вязкого трения
alpha_values = np.linspace(0, np.pi/2, 100)  # Диапазон начальных отклонений

T_values = 2 * np.pi / np.sqrt(k) * np.ones_like(alpha_values)

plt.figure(figsize=(8, 6))
plt.plot(alpha_values, T_values, label='T(α)')
plt.xlabel('Начальное отклонение α (рад)')
plt.ylabel('Время затухания T (сек)')
plt.title('Зависимость времени затухания от начального отклонения маятника')
plt.legend()
plt.grid(True)
plt.show()
```

# Пункт 4: Зависимость периода колебания маятника от коэффициента сопротивления и определение, при каком коэффициенте сопротивления колебания маятника отсутствуют

```python
k = 10.0  # Жесткость пружины
M = 1.0   # Масса маятника
c_values = np.linspace(0, 20, 100)  # Диапазон коэффициентов сопротивления

# Рассчитываем период колебания T для каждого коэффициента сопротивления c
omega0 = np.sqrt(k / M)
T_values = 2 * np.pi / np.sqrt(omega0**2 - (c_values / (2 * M))**2)


c_crit = 2 * np.sqrt(M * k)

plt.figure(figsize=(8, 6))
plt.plot(c_values, T_values, label='T(c)')
plt.axvline(x=c_crit, color='red', linestyle='--', label='c_crit')
plt.xlabel('Коэффициент сопротивления c')
plt.ylabel('Период колебания T (сек)')
plt.title('Зависимость периода колебания от коэффициента сопротивления')
plt.legend()
plt.grid(True)
plt.show()
```

# Нарисовать фазовую диаграмму для исходных значений

```python
M = 1.0  # Масса маятника
k = 10.0  # Жесткость пружины
c = 0.5  # Коэффициент вязкого трения
alpha0 = 0.1  # Начальное отклонение (рад)
v0 = 0.0  # Начальная скорость

# Функция, описывающая систему диффуров
def system(state, t):
    x, v = state
    dxdt = v
    dvdt = - (c / M) * v - (k / M) * x
    return [dxdt, dvdt]


t = np.linspace(0, 10, 1000)  # Временной интервал
t0 = 0  
state0 = [alpha0, v0]  # Начальное состояние

# Интегрирование системы дифференциальных уравнений
states = odeint(system, state0, t)

# Извлечение координат и скоростей
x_values, v_values = states[:, 0], states[:, 1]

plt.figure(figsize=(8, 6))
plt.plot(x_values, v_values)
plt.xlabel('Координата x')
plt.ylabel('Скорость v')
plt.title('Фазовая диаграмма для исходных значений')
plt.grid(True)
plt.show()
```
